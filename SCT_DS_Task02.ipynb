```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('train.csv')
df.head(3)
```

```python
# Identify missing values and drop irrelevant information
df = df.drop(columns=['Cabin','PassengerId','Name','Ticket'])
print(df.isnull().sum())
```

```python
# Fill in the missing values 
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print(df.isnull().sum())
df.head(5)
```

```python
#Plot 1
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived', hue='Survived', palette="pastel", legend=False)
plt.title('SURVIVAL COUNT')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()

#Plot 2
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sex', hue='Survived', palette="pastel")
plt.title('SURVIVAL COUNT BY GENDER')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

#Plot 3 
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='pastel')
plt.title('SURVIVAL COUNT BY PASSENGER CLASS')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()  

#Polt 4
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Embarked', hue='Survived', palette='pastel')
plt.title('SURVIVAL COUNT BY EMBARKMENT POINT')
plt.xlabel('Embarked')
plt.ylabel('Passenger Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

#PLOT 5
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', kde=True, palette='pastel')
plt.title('SURVIVAL COUNT BY AGE')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

#PLOT 6
plt.figure(figsize=(6,4))
corr = df[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('CORRELATION HEATMAP')
plt.show()  

#PLOT 7
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='pastel')
plt.suptitle('PAIRPLOT OF KEY FEATURES', y=1.05)
plt.show()

```
